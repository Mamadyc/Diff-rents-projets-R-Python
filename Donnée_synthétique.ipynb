# Donnée synthétique et quelques paramètres (calculs)

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors

def create_synthetic_data(X, n_components=2, n_neighbors=5):
    # Calcul des moyennes des colonnes
    mu = np.mean(X, axis=0)

    # Réduction de la dimensionnalité avec PCA
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X)

    # Recherche des k plus proches voisins
    nn = NearestNeighbors(n_neighbors=n_neighbors)
    nn.fit(X_pca)
    distances, indices = nn.kneighbors(X_pca)

    # Création de nouvelles coordonnées basées sur les voisins pondérés
    new_coord = np.zeros_like(X_pca)
    for i in range(X_pca.shape[0]):
        inv_distances = 1. / np.maximum(1e-9, distances[i])
        weights = inv_distances * np.random.exponential(size=n_neighbors) * 0.5**np.random.permutation(n_neighbors)
        weights /= np.sum(weights)
        new_coord[i] = np.average(X_pca[indices[i]], axis=0, weights=weights)

    # Retour dans l'espace d'origine
    X_synthetic = np.dot(new_coord, pca.components_) + mu

    return X_synthetic
```

# Comparaison des colonnes des deux bases des données synthétiques et originale


1. Comparaison visuelle avec des sous figures



```python
import matplotlib.pyplot as plt

# Comparaison visuelle entre les données initiales et synthétiques
def plot_comparison(original_data, synthetic_data):
    num_columns = len(original_data.columns)
    fig, axes = plt.subplots(num_columns, 2, figsize=(12, num_columns*3))
    for i, column in enumerate(original_data.columns):
        # Plot original data
        axes[i, 0].hist(original_data[column], bins=20, color='blue', alpha=0.5, label='Original')
        axes[i, 0].set_title(f'Original {column}')
        axes[i, 0].legend()

        # Plot synthetic data
        axes[i, 1].hist(synthetic_data[column], bins=20, color='orange', alpha=0.5, label='Synthetic')
        axes[i, 1].set_title(f'Synthetic {column}')
        axes[i, 1].legend()

    plt.tight_layout()
    plt.show()

# Utilisation de la fonction pour comparer les données
plot_comparison(df, df_synthetic)
```
